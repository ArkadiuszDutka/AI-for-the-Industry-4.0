```python
data_air01 = np.loadtxt("/content/drive/My Drive/sensor01_air.txt", delimiter=",", skiprows=1)
data_izopropanol01 = np.loadtxt("/content/drive/My Drive/sensor01_izopropanol.txt", delimiter=",", skiprows=1)
data_water01 = np.loadtxt("/content/drive/My Drive/sensor01_water.txt", delimiter=",", skiprows=1)

# Get time and signal values
time_air01 = data_air01[:, 0]
signal_air01 = data_air01[:, 1]
time_izopropanol01 = data_izopropanol01[:, 0]
signal_izopropanol01 = data_izopropanol01[:, 1]
time_water01 = data_water01[:, 0]
signal_water01 = data_water01[:, 1]

# Perform FFT
freq_air01 = np.fft.fftfreq(signal_air01.size, d=time_air01[1]-time_air01[0])
fft_air01 = np.fft.fft(signal_air01)
freq_izopropanol01 = np.fft.fftfreq(signal_izopropanol01.size, d=time_izopropanol01[1]-time_izopropanol01[0])
fft_izopropanol01 = np.fft.fft(signal_izopropanol01)
freq_water01 = np.fft.fftfreq(signal_water01.size, d=time_water01[1]-time_water01[0])
fft_water01 = np.fft.fft(signal_water01)

# Plot FFT
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.plot(freq_air01, np.abs(fft_air01))
plt.title("FFT - Air01")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.subplot(132)
plt.plot(freq_izopropanol01, np.abs(fft_izopropanol01))
plt.title("FFT - Izopropanol01")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.subplot(133)
plt.plot(freq_water01, np.abs(fft_water01))
plt.title("FFT - Water01")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Load data
data_air01 = np.loadtxt('/content/drive/My Drive/sensor01_air.txt', delimiter=',', skiprows=1)
data_izopropanol01 = np.loadtxt('/content/drive/My Drive/sensor01_izopropanol.txt', delimiter=',', skiprows=1)
data_water01 = np.loadtxt('/content/drive/My Drive/sensor01_water.txt', delimiter=',', skiprows=1)

# Get time and signal values
time_air01 = data_air01[:, 0]
signal_air01 = data_air01[:, 1]
time_izopropanol01 = data_izopropanol01[:, 0]
signal_izopropanol01 = data_izopropanol01[:, 1]
time_water01 = data_water01[:, 0]
signal_water01 = data_water01[:, 1]

# Perform linear regression
from sklearn.linear_model import LinearRegression

model_air01 = LinearRegression().fit(time_air01.reshape(-1, 1), signal_air01.reshape(-1, 1))
model_izopropanol01 = LinearRegression().fit(time_izopropanol01.reshape(-1, 1), signal_izopropanol01.reshape(-1, 1))
model_water01 = LinearRegression().fit(time_water01.reshape(-1, 1), signal_water01.reshape(-1, 1))

# Get regression coefficients
coeffs_air01 = model_air01.coef_[0][0]
intercept_air01 = model_air01.intercept_[0]
coeffs_izopropanol01 = model_izopropanol01.coef_[0][0]
intercept_izopropanol01 = model_izopropanol01.intercept_[0]
coeffs_water01 = model_water01.coef_[0][0]
intercept_water01 = model_water01.intercept_[0]

# Print regression coefficients
print("Air01 - Coeffs: {}, Intercept: {}".format(coeffs_air01, intercept_air01))
print("Izopropanol01 - Coeffs: {}, Intercept: {}".format(coeffs_izopropanol01, intercept_izopropanol01))
print("Water01 - Coeffs: {}, Intercept: {}".format(coeffs_water01, intercept_water01))
```
