```python
df_air01 = pd.read_csv('/content/drive/My Drive/sensor01_air.txt', header=None, names=['Time_air01', 'Value_air01'])

# Read the contents of the second file into a pandas DataFrame
df_izopropanol01 = pd.read_csv('/content/drive/My Drive/sensor01_izopropanol.txt', header=None, names=['Time_izopropanol01', 'Value_izopropanol01'])

df_water01 = pd.read_csv('/content/drive/My Drive/sensor01_water.txt', header=None, names=['Time_water01', 'Value_water01'])

# Create a line plot of the 'Value' column for both data series
plt.plot(df_air01['Value_air01'], label='Air')
plt.plot(df_izopropanol01['Value_izopropanol01'], label='Izopropanol')
plt.plot(df_water01['Value_water01'], label='Water')


# Add a legend to the plot
plt.legend()

# Add axis labels and a title to the plot
plt.xlabel(' [nm]')
plt.ylabel('P\Pref [dB]')
plt.title('Sensor Data')

# Display the plot
plt.show()
```
